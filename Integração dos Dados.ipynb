```python
import pandas as pd
```

```python
chuva_df = pd.read_csv('PivotChuva_202309261223.csv', parse_dates=['Data'])
chuva_df.drop_duplicates(subset='Data', inplace=True)
chuva_df
```

```python
cota_df = pd.read_csv('PivotCota_202309261222.csv', parse_dates=['Data'])
cota_df.drop_duplicates(subset='Data', inplace=True)
cota_df
```

```python
vazao_df = pd.read_csv('PivotVazao_202309261223.csv', parse_dates=['Data'])
vazao_df.drop_duplicates(subset='Data', inplace=True)
vazao_df
```

```python
# Interseção de cota_df e chuva_df por data
df = pd.merge(cota_df, chuva_df, on='Data', how='inner', suffixes=('_cota', '_chuva'))
```

```python
# Une df e vazao_df por data
df = pd.merge(df, vazao_df, on='Data', how='inner')
```

```python
df
```

```python
# Renomeia NivelConsistencia para NivelConsistencia_vazao
df.rename(columns={'NivelConsistencia':'NivelConsistencia_vazao', 'EstacaoCodigo':'EstacaoCodigo_vazao'}, inplace=True)
```

```python
df.head()
```

```python
# Dropa colunas desnecessárias
df.drop(columns=['RegistroID_cota','CotaID','RegistroID_chuva','ChuvaID'], inplace=True)
df.drop(columns=['RegistroID','VazaoID'], inplace=True)
```

```python
df.head()
```

```python
# Rearranjando as colunas
df = df[['Data','EstacaoCodigo_chuva','NivelConsistencia_chuva','Chuva','EstacaoCodigo_cota','NivelConsistencia_cota','Cota','EstacaoCodigo_vazao','NivelConsistencia_vazao','Vazao']]
df.head()
```

```python
# Salva em um csv
df.to_csv('chuva_vazao_cota_CAIANA.csv', date_format='%Y-%m-%d', index=False)
```

## Adicionando dados de temperatura

```python
temperatura_df = pd.read_csv('temperatura_CAPARAO.csv', parse_dates=['Data Medicao'])
```

```python
temperatura_df
```

```python
temperatura_df.rename(columns={'Data Medicao':'Data', 'TEMPERATURA MEDIA COMPENSADA':'temperatura_media_compensada'}, inplace=True)
```

```python
# Reordena por data de forma decrescente
temperatura_df.sort_values(by='Data', ascending=False, inplace=True)
```

```python
# Une df e vazao_df por data
df = pd.merge(df, temperatura_df, on='Data', how='inner')
```

```python
# Salva em um csv
df.to_csv('temperatura_chuva_vazao_cota_CAIANA.csv', date_format='%Y-%m-%d', index=False)
```
